We'll be working with a dataset on the job outcomes of USA students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:
- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.

# Introduction 
Before we start creating data visualizations, let's import the libraries we need and remove rows containing null values.

```python
import pandas as pd
import matplotlib as plt

# display plots inline
% matplotlib inline    

recent_grads = pd.read_csv('recent-grads.csv')

print(recent_grads.iloc[0])    #return first row as a table
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

Drop rows with missing values:

```python
raw_data_count = len(recent_grads)    # number of rows before cleaning
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)    # number of rows after cleaning
print(raw_data_count)
print(cleaned_data_count)
```

Notice that only one row contained missing values and was dropped.

Use the plots to explore the following questions:
- Do students in more popular majors make more money?
No. From the 'sample size vs. median' scatter plot, it doesn't appear that majors with a bigger sample size have a higher median. They tend to have an average median. 

- Do students that majored in subjects that were majority female make more money?
No, the opposite, students that majored in subjects that had a majority of female students, make less money. This can be seen by comparing the 'men vs. median' scatterplot with the 'women vs. median' scatterplot. To make it more clear, I created an extra scatterplot 'ShareWomen vs. Median'. This shows that the higher the share of women, the lower the median.

- Is there any link between the number of full-time employees and median salary?
No, in the 'full time vs. median' scatter plot, a relation between these does not show.

# Pandas, Bar Plots

Use bar plots to compare the percentages of women (ShareWomen) from the first ten rows and last ten rows of the recent_grads dataframe.

```python
# first ten rows:
recent_grads[:10].plot.bar(x='Major', y='ShareWomen')
```

```python
# last ten rows:
recent_grads[163:].plot.bar(x='Major', y='ShareWomen')
```

The whole dataset is ordered by median salary. This shows that in the higher salary-majors, women are underrepresented, with a few exeptions. In the lower salary-majors, women are the vast majority, with every major having more than 60% women (and most a lot more).
